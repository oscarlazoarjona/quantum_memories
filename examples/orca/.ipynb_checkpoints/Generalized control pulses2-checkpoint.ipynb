```python
from quantum_memories.misc import num_integral
def square_normalized(t, t0, tau, norm=1):
    f = np.where((t-t0)/tau >=-0.5, 1.0, 0.0)*np.where((t-t0)/tau <=0.5, 1.0, 0.0)
    f = np.sqrt(norm)*f
    return f

def Omega_generalized(t, t0, tau, E, m=oo):
    if m == oo:
        sig = tau/2
        #f = np.sqrt(E)*square(t-t0, 2*sig)/2/sig
        f = square_normalized(t, t0, tau, E)
    else:
        sig = tau/2.0/np.sqrt(2.0*np.log(2.0))**(1.0/m)
        f = np.sqrt(E)/np.sqrt(2*sig*gamma_function(1+1.0/2/m))
        f = f*np.exp(-((t-t0)/sig)**(2*m)/2.0)
    f = f/np.sqrt(num_integral(t, np.abs(f)**2)/E)
    return f
```

```python
plt.figure(figsize=(15, 8))
ts = np.linspace(-50, 50, 10000)
tau1s = 1.0
t0s = 0.0
Es = 1.0
f01 = Omega_generalized(ts, t0s, tau1s, Es, 1)
f02 = Omega_generalized(ts, t0s, tau1s, Es, 2)
f10 = Omega_generalized(ts, t0s, tau1s, Es, 10)
f99 = Omega_generalized(ts, t0s, tau1s, Es, 99)
foo = Omega_generalized(ts, t0s, tau1s, Es, oo)

from quantum_memories.misc import num_integral
print num_integral(ts, np.abs(f01)**2)
print num_integral(ts, np.abs(f02)**2)
print num_integral(ts, np.abs(f10)**2)
print num_integral(ts, np.abs(f99)**2)
print num_integral(ts, np.abs(foo)**2)

plt.plot(ts, f01)
plt.plot(ts, f02)
plt.plot(ts, f10)
plt.plot(ts, f99)
plt.plot(ts, foo, "k-")
plt.xlim(t0s-2.5*tau1s/2, t0s+2.5*tau1s/2)
```

```python
def ffftfreq(t):
    r"""Calculate the angular frequency axis for a given time axis."""
    dt = t[1]-t[0]
    nu = np.fft.fftshift(np.fft.fftfreq(t.size, dt))
    return nu


def ffftfft(f, t):
    r"""Calculate the Fourier transform."""
    dt = t[1]-t[0]
    return np.fft.fftshift(np.fft.fft(np.fft.ifftshift(f)))*dt


def iffftfft(f, nu):
    r"""Calculate the inverse Fourier transform."""
    Deltanu = nu[-1]-nu[0]
    return np.fft.fftshift(np.fft.ifft(np.fft.ifftshift(f)))*Deltanu
```
