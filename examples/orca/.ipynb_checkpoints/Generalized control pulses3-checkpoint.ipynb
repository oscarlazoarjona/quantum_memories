```python
from sympy import symbols, init_printing, exp, pi, oo, integrate, simplify, gamma, sqrt, solve, fourier_transform
from sympy import inverse_fourier_transform, sin, Piecewise, I, Equality
from matplotlib import pyplot as plt
from scipy.special import gamma as gamma_function
import numpy as np
init_printing()
%matplotlib inline
```

```python
t, t0, nu = symbols("t, t0, nu", real=True)
sigma1, tau1, E, E0 = symbols("sigma1, tau1, E, E0", positive=True)
m = symbols("m", integer=True, positive=True)

Et = sqrt(E)*exp(-((t-t0)/sigma1)**(2*m)/2)/sqrt(2*sigma1*gamma(1+1/m/2))
Et
```

```python
Et.subs(m, 1)
```

```python
simplify(integrate(abs(Et)**2, (t, -oo, oo)))
```

We calculate the time width.

```python
eq = abs(Et.subs(t, t0))/2 - abs(Et)
eq = abs(Et.subs(t, t0))**2/2 - abs(Et)**2
hwhm = solve(eq, t)[0]-t0
fwhm = 2*hwhm
ss_fwhm = {tau1: fwhm}
ss_fwhm_inv = {sigma1: sigma1*tau1/fwhm}
ss_fwhm, ss_fwhm_inv
```

```python
Powt = abs(Et)**2
Et, Powt
```

```python
Powt.subs(t, t0), Powt.subs(t, t0+fwhm/2)
```

```python
from quantum_memories.misc import num_integral
def square_normalized(t, t0, tau, norm=1):
    f = np.where((t-t0)/tau >=-0.5, 1.0, 0.0)*np.where((t-t0)/tau <=0.5, 1.0, 0.0)
    f = np.sqrt(norm)*f
    return f

def Omega_generalized(t, t0, tau, E, m=oo):
    if m == oo:
        sig = tau/2
        #f = np.sqrt(E)*square(t-t0, 2*sig)/2/sig
        f = square_normalized(t, t0, tau, E)
    else:
        sig = tau/2.0/np.sqrt(2.0*np.log(2.0))**(1.0/m)
        sig = tau/2/np.log(2)**(1.0/2/m)
        f = np.sqrt(E)/np.sqrt(2*sig*gamma_function(1+1.0/2/m))
        f = f*np.exp(-((t-t0)/sig)**(2*m)/2.0)
    f = f/np.sqrt(num_integral(t, np.abs(f)**2)/E)
    return f
```

```python
ts = np.linspace(-50, 50, 10000)
tau1s = 1.0
t0s = 0.0
Es = 1.0
f01 = Omega_generalized(ts, t0s, tau1s, Es, 1)
f02 = Omega_generalized(ts, t0s, tau1s, Es, 2)
f10 = Omega_generalized(ts, t0s, tau1s, Es, 10)
f99 = Omega_generalized(ts, t0s, tau1s, Es, 99)
foo = Omega_generalized(ts, t0s, tau1s, Es, oo)

from quantum_memories.misc import num_integral
print num_integral(ts, np.abs(f01)**2)
print num_integral(ts, np.abs(f02)**2)
print num_integral(ts, np.abs(f10)**2)
print num_integral(ts, np.abs(f99)**2)
print num_integral(ts, np.abs(foo)**2)

plt.figure(figsize=(13, 10))

plt.subplot(211)
plt.plot(ts, np.abs(f01))
plt.plot(ts, np.abs(f02))
plt.plot(ts, np.abs(f10))
plt.plot(ts, np.abs(f99))
plt.plot(ts, np.abs(foo), "k-")
plt.xlim(t0s-2.5*tau1s/2, t0s+2.5*tau1s/2)
plt.xlabel(r"$t$", fontsize=15)
plt.ylabel(r"$|S|$", fontsize=15)

plt.subplot(212)
plt.plot(ts, f01**2)
plt.plot(ts, f02**2)
plt.plot(ts, f10**2)
plt.plot(ts, f99**2)
plt.plot(ts, foo**2, "k-")
plt.xlim(t0s-2.5*tau1s/2, t0s+2.5*tau1s/2)
plt.xlabel(r"$t$", fontsize=15)
plt.ylabel(r"$|S|^2$", fontsize=15)
```

The bandwidth for $m=1$

```python
Pow01hat = abs(E01hat)**2

bandwidth01 = solve(Pow01hat.subs(nu, 0)/2-Pow01hat, nu)
bandwidth01 = bandwidth01[3]-bandwidth01[2]
bandwidth01 = bandwidth01.subs(ss_fwhm_inv).subs(m, 1)
bandwidth01 = simplify(bandwidth01)
bandwidth01, bandwidth01.subs(tau1, 1).n()
```

The bandwidth for $m=\infty$

```python
Powoohat = abs(Eoohat)**2
s = symbols("s", positive=True)
eq = Equality(Powoohat.limit(nu, 0)/2, Powoohat)
ff = (pi*nu)**2/E*tau1
eq = Equality(eq.lhs*ff, eq.rhs*ff)
eq = Equality(sqrt(eq.lhs), sqrt(eq.rhs))

eq = eq.subs(abs(nu)*tau1*pi, s)
eq = eq.subs(nu*tau1*pi, s)
eq = eq.subs(abs(sin(s)), sin(s))

eq
```

```python
def ffftfreq(t):
    r"""Calculate the angular frequency axis for a given time axis."""
    dt = t[1]-t[0]
    nu = np.fft.fftshift(np.fft.fftfreq(t.size, dt))
    return nu


def ffftfft(f, t):
    r"""Calculate the Fourier transform."""
    dt = t[1]-t[0]
    return np.fft.fftshift(np.fft.fft(np.fft.ifftshift(f)))*dt


def iffftfft(f, nu):
    r"""Calculate the inverse Fourier transform."""
    Deltanu = nu[-1]-nu[0]
    return np.fft.fftshift(np.fft.ifft(np.fft.ifftshift(f)))*Deltanu
```

```python
nus = ffftfreq(ts)
foo_hat = ffftfft(foo, ts)
sigma1s = tau1s/(2*np.sqrt(np.log(2)))

E01hats = np.sqrt(2*np.sqrt(np.pi)*Es*sigma1s)*np.exp(-2*(np.pi*nus*sigma1s)**2)
Eoohats = np.sqrt(Es/tau1s)/nus/np.pi*np.sin(np.pi*nus*tau1s)

E01hats = ffftfft(f01, ts)
Eoohats = ffftfft(foo, ts)

# Widths for amplitude and time.
f01max_ampt = np.sqrt(Es/sigma1s/np.sqrt(np.pi))
foomax_ampt = np.sqrt(Es/tau1s)
fwhm01_ampt = np.sqrt(2.0)*tau1s
fwhmoo_ampt = tau1s

# Widths for power and time.
f01max_powt = f01max_ampt**2
foomax_powt = foomax_ampt**2
fwhm01_powt = tau1s
fwhmoo_powt = tau1s

# Withs for power and frequency.
fmax01 = 2*np.sqrt(np.pi)*Es*sigma1s
fmaxoo = Es*tau1s
fwhms01 = float(bandwidth01.subs(tau1, tau1s))
fwhmsoo = float(bandwidthoo.subs(tau1, tau1s))
#fwhmsoo = 2*root_hwhm/np.pi/tau1s

plt.figure(figsize=(15, 8))

plt.subplot(221)
plt.plot(ts, np.abs(f01), "b-")
plt.plot(ts, np.abs(foo), "r-")
plt.plot([t0s-fwhm01_ampt/2, t0s+fwhm01_ampt/2], [f01max_powt/2, f01max_powt/2], "b:")
plt.plot([t0s-fwhmoo_ampt/2, t0s+fwhmoo_ampt/2], [foomax_powt/2, foomax_powt/2], "r:")
#plt.plot([t0s-tau1s/2, t0s+tau1s/2], [f01max_ampt/2, f01max_ampt/2], "b:")
#plt.plot([t0s-tau1s/2, t0s+tau1s/2], [foomax_ampt/2, foomax_ampt/2], "r:")
plt.xlabel(r"$t$", fontsize=15)
plt.ylabel(r"$|S|$", fontsize=15)
plt.xlim(t0s-tau1s*2, t0s+tau1s*2)

plt.subplot(222)
plt.plot(nus, np.abs(E01hats), "b-")
plt.plot(nus, np.abs(Eoohats), "r-")
plt.xlabel(r"$\nu$", fontsize=15)
plt.ylabel(r"$|\hat{S}|$", fontsize=15)
plt.xlim(-3*fwhmsoo, 3*fwhmsoo)

plt.subplot(223)
plt.plot(ts, np.abs(f01)**2, "b-")
plt.plot(ts, np.abs(foo)**2, "r-")
plt.plot([t0s-fwhm01_powt/2, t0s+fwhm01_powt/2], [f01max_powt/2, f01max_powt/2], "b:")
plt.plot([t0s-fwhmoo_powt/2, t0s+fwhmoo_powt/2], [foomax_powt/2, foomax_powt/2], "r:")

plt.xlabel(r"$t$", fontsize=15)
plt.ylabel(r"$|S|^2$", fontsize=15)
plt.xlim(t0s-tau1s*2, t0s+tau1s*2)

plt.subplot(224)
plt.plot(nus, np.abs(E01hats)**2, "b-")
plt.plot(nus, np.abs(Eoohats)**2, "r-")
plt.plot([-fwhms01/2, fwhms01/2], [fmax01/2, fmax01/2], "b:")
plt.plot([-fwhmsoo/2, fwhmsoo/2], [fmaxoo/2, fmaxoo/2], "r:")
plt.xlabel(r"$\nu$", fontsize=15)
plt.ylabel(r"$|\hat{S}|^2$", fontsize=15)
plt.xlim(-3*fwhmsoo, 3*fwhmsoo)

plt.savefig("/home/oscar/Desktop/a2.png", bbox_inches="tight")
```
